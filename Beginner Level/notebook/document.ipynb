```python
### pdf file loading
from langchain_community.document_loaders import PyPDFLoader

### Data Loading
pdf_path = Path('data/PrashantJoshi_CV_DS.pdf')
if not pdf_path.exists():
    pdf_path = (Path.cwd().resolve().parent / 'data' / 'PrashantJoshi_CV_DS.pdf').resolve()

if not pdf_path.exists():
    raise FileNotFoundError(f"Document not found: {path}")

loader = PyPDFLoader(str(pdf_path))
pdf_document = loader.load()
pdf_document
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
text_chunks = text_splitter.split_documents(pdf_document)
text_chunks
```

Vector Embedding Techniques and Vector Stores

```python
#### for creating vectors
from langchain_community.embeddings import OpenAIEmbeddings 

### for storing vectors in a vector database, we can use chromadb, pinecone, 
# weaviate, etc. Here we will use chromadb for simplicity
from langchain_community.vectorstores import Chroma

chroma_db = Chroma.from_documents(text_chunks, OpenAIEmbeddings())
```

```python
### Vector Db

query = "Who is Prashant?"
result = chroma_db.similarity_search(query)
result
```

```python
#### FAISS Vector Store
from langchain_community.vectorstores import FAISS 
faiss_db = FAISS.from_documents(text_chunks, OpenAIEmbeddings())
```

```python
query = "Who is Prashant?"
result = faiss_db.similarity_search(query)
result
```
